```python
import numpy as np
import pandas as pa
from matplotlib import pyplot as plt
import datetime
%matplotlib inline
```

```python
df = pa.read_hdf('data.hdf')
```

```python
plt.plot(df.loc[:].average_counts,'.')
df = df.loc[:]
df.keys()
```

```python
df_Sub = df.groupby(['phase_pi2_2','tau_2', 'sweeps']).average_counts.agg('mean')
max_sweep = max(df.sweeps.unique())
```

```python
# nuclear baths
Bfield = 0.0881 # Tesla
gamma_C = 10.7 #T/MHz
gamma_Si = -8.5 #T/MHz
pi2_duration = 0.05
larmor_C = Bfield*gamma_C
larmor_Si = Bfield*gamma_Si
bath_C = 1/(2*larmor_C)-pi2_duration
bath_Si = abs(1/(2*larmor_Si)-pi2_duration) # is is physical to use absolute value?

print(bath_C)
print(bath_Si)
```

```python
%matplotlib notebook
tau = df.tau_2.unique()
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10,5))
y1 = np.zeros(len(tau))
y2 = np.zeros(len(tau))
for sweep in df.sweeps.unique():  
    if sweep != 27: #exclude last sweep which was bad
        y1+= np.array(df_Sub.loc[0, :, sweep])
        y2+= np.array(df_Sub.loc[180, :, sweep])
y = y2-y1
y = y/max(y)

ax1.plot(tau,y,'.-')
ax2.plot(tau,y,'.-')
#ax1.plot(tau,y_2,'.-')
#ax2.plot(tau,y_2,'.-')
ax2.vlines(bath_C,min(y),max(y))
ax2.vlines(bath_Si,min(y),.5*max(y))
ax2.text(bath_C+0.01, .8*max(y),'C bath')
ax2.text(bath_Si+0.01, .5*max(y),'Si bath')
ax1.set_xlim([0,10])
ax2.set_xlim([0,3])
plt.tight_layout()
#plt.show()
```
